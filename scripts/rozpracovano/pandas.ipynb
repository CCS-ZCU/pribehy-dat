```python
def determine_long_words(val):
  if len(val) > 3:
    output = val + " - LONG!"
  else:
    output = val + " - SHORT!"
  return output
```

```python
calif = pd.read_csv("sample_data/california_housing_train.csv")
```

```python
def det_pop_groups(val):
  if val > 1000:
    group = "big"
  elif 1000 >= val > 500:
      group = "middle"
  else:
      group = "small"
  return group
```

```python
det_pop_groups(1200)
```

```python
calif["pop_groups"] = calif["population"].apply(det_pop_groups)
```

```python
calif["pop_groups"].hist()
```
